```python
unfetched = pd.read_table('unfetched_pmids.tsv', header=None)

unfetched['doi'] = unfetched[0].map(id_dict)
```

```python
sci_list = unfetched['doi'].tolist()
```

```python
sci_list
```

```python
def download_multi_papers(source):
    for x in source:
        scihub_download(x, paper_type="doi", out="./fetched_pdfs/")
```

```python
download_multi_papers(sci_list)
```
